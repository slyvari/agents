```python
from dotenv import load_dotenv
import os
from openai import OpenAI
from IPython.display import Markdown, display

# Load .env file
load_dotenv(override=True)

# Read OpenAI API key
openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key:
    openai_api_key = openai_api_key.strip().replace('"', '').replace("'", "")
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    raise ValueError("OPENAI_API_KEY not set")

# Create OpenAI client
openai_client = OpenAI(api_key=openai_api_key)
```

```python
question1 = """
You are a helpful assistant.

Pick one business area that might be worth exploring for an Agentic AI opportunity.
Give only one business area and explain in 2-3 lines why it is suitable.
"""

messages1 = [
    {"role": "user", "content": question1}
]

response1 = openai_client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=messages1
)

business_area = response1.choices[0].message.content

print("BUSINESS AREA:")
print(business_area)
```

```python
question2 = f"""
You are a helpful assistant.

The selected business area is:

{business_area}

Now present one major pain point in this industry.
The pain point should be challenging and suitable for an Agentic AI solution.
Explain it in simple terms.
"""

messages2 = [
    {"role": "user", "content": question2}
]

response2 = openai_client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=messages2
)

pain_point = response2.choices[0].message.content

print("PAIN POINT:")
print(pain_point)
```

```python
question3 = f"""
You are a helpful assistant.

Business area:
{business_area}

Pain point:
{pain_point}

Now propose an Agentic AI solution.
"""

messages3 = [
    {"role": "user", "content": question3}
]

response3 = openai_client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=messages3
)

agentic_solution = response3.choices[0].message.content

print("AGENTIC AI SOLUTION:")
print(agentic_solution)
display(Markdown(agentic_solution))
```
