**Theme:** Agentic AI opportunity in the **driving school** industry.

This notebook completes the Week 1 exercise:
1. Ask the LLM to pick a business area worth exploring
2. Ask for a pain-point in that industry
3. Ask for an Agentic AI solution to that pain-point

```python
from dotenv import load_dotenv
import os
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)

openai_api_key = os.getenv("OPENAI_API_KEY")
if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")

openai = OpenAI()
```

## Step 1 — Pick a business area

We steer the LLM toward driving schools, but still let it suggest a specific niche.

```python
# First create the messages:

messages = [{
    "role": "user",
    "content": (
        "Pick one business area within the driving school industry that might be worth "
        "exploring for an Agentic AI opportunity. Respond in 2-3 sentences."
    ),
}]

# Then make the first call:

response = openai.chat.completions.create(model="gpt-5.4-nano", messages=messages)

# Then read the business area:

business_area = response.choices[0].message.content
print(business_area)
```

## Step 2 — Identify a pain-point

Include the business area from step 1 in the next message.

```python
messages = [{
    "role": "user",
    "content": (
        f"In this business area: {business_area}\n\n"
        "Describe one challenging pain-point that might be ripe for an Agentic AI solution. "
        "Be specific about who feels the pain and why it is hard to solve today."
    ),
}]

response = openai.chat.completions.create(model="gpt-5.4-mini", messages=messages)

pain_point = response.choices[0].message.content
display(Markdown(pain_point))
```

## Step 3 — Propose an Agentic AI solution

```python
messages = [{
    "role": "user",
    "content": (
        f"Business area: {business_area}\n\n"
        f"Pain-point: {pain_point}\n\n"
        "Propose an Agentic AI solution. Describe the agents involved, what each one does, "
        "and how they collaborate to solve the pain-point."
    ),
}]

response = openai.chat.completions.create(model="gpt-5.4-mini", messages=messages)

agentic_solution = response.choices[0].message.content
display(Markdown(agentic_solution))
```
